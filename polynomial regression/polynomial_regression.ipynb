```python
import pandas as pd
import matplotlib.pyplot as mpt
```

```python
data_set = pd.read_csv("Position_Salaries.csv")
```

```python
data_set
```

```python
temp = data_set[["Level","Salary"]]
temp.corr()
```

```python
x = data_set[["Level"]].values
y = data_set[["Salary"]].values
```

```python
print(x)
print(y)
```

```python
mpt.scatter(x,y, color = "green")
mpt.title("level vs salary")
mpt.xlabel("level")
mpt.ylabel("salary")
mpt.show()
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x,y)
```

```python
pred = lr.predict(x)
print(pred)
```

```python
lr.score(x,y)*100
```

```python
mpt.scatter(x,y,color= "blue")
mpt.plot(x,pred,color = "red")
mpt.title("level vs salary using simple regression")
mpt.xlabel("level")
mpt.ylabel("salary")
mpt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly_reg = PolynomialFeatures(degree = 2)
x_poly = poly_reg.fit_transform(x)
```

```python
lr_reg = LinearRegression()
lr_reg.fit(x_poly,y)
```

```python
lr_reg.score(x_poly,y)*100
```

```python
poly_pred = lr_reg.predict(x_poly)
```

```python
poly_pred
```

```python
mpt.scatter(x,y, color = "green")
mpt.plot(x,poly_pred, color = "red")
mpt.title("level vs salary using polynomial regression with degree 2")
mpt.xlabel("level")
mpt.ylabel("salary")
mpt.show()
```

```python
poly_reg2 = PolynomialFeatures(degree = 3)
x_poly2 = poly_reg2.fit_transform(x)
lr_reg2 = LinearRegression()
lr_reg2.fit(x_poly2,y)
```

```python
lr_reg2.score(x_poly2,y)*100
```

```python
mpt.scatter(x,y, color = "blue")
mpt.plot(x,lr_reg2.predict(x_poly2), color = "red")
mpt.title("level vs salary using polynomial regression with degree 3")
mpt.xlabel("level")
mpt.ylabel("salary")
mpt.show()
```
