```python
import pandas as pd
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
df = pd.DataFrame(iris.data, columns = iris.feature_names)
df
```

```python
df['target'] = iris.target
df.head()
```

```python
df[df.target == 2].head()
```

```python
df[df.target == 1].head()
```

```python
df['flower_name'] = df.target.apply(lambda x: iris.target_names[x])
df.head()
```

```python
df0 = df[df.target == 0]
df1 = df[df.target == 1]
df2 = df[df.target == 2]
```

```python
import matplotlib.pyplot as mtp
```

```python
mtp.xlabel('saple length (cm)')
mtp.ylabel('saple width (cm)')
mtp.scatter(df0['sepal length (cm)'],df0['sepal width (cm)'], marker = '+', color = 'green')
mtp.scatter(df1['sepal length (cm)'],df1['sepal width (cm)'], marker = '.', color = "blue")
```

```python
mtp.xlabel('petal length (cm)')
mtp.ylabel('petal width (cm)')
mtp.scatter(df0['petal length (cm)'],df0['petal width (cm)'], marker = '+', color = 'green')
mtp.scatter(df1['petal length (cm)'],df1['petal width (cm)'], marker = '.', color = "blue")
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = df.drop(['target','flower_name'], axis = 'columns')
x.head()
```

```python
y = df.target
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2)
```

```python
from sklearn.svm import SVC
```

```python
model = SVC()
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)*100
```

```python
pred = model.predict(x_test)
```
