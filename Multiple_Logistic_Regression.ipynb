```python
import pandas as pd
```

```python
df = pd.read_csv("Iris_Dataset.csv")
```

```python
df
```

```python
df['Species'].unique()
```

```python
df['Species'].replace({'Iris-virginica':'1', 'Iris-setosa':'2', 'Iris-versicolor':'3'},inplace = True)
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(df[["SepalLengthCm","SepalWidthCm","PetalLengthCm","PetalWidthCm"]], df['Species'], test_size = 0.2)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr.predict(x_test)
```

```python
lr.score(x_test,y_test)
```
