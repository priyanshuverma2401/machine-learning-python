```python
df = pd.DataFrame(digits.data)
```

```python
df.head()
```

```python
df['target'] = digits.target
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = df.drop(['target'], axis = 'columns')
y = df.target
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2)
```

```python
print('lwngth of x-train is', len(x_train))
print('lwngth of y-train is', len(y_train))
```

```python
print('length of x_test is ',len(x_test))
print('length of y_test is ',len(y_test))
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators = 10)
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)*100
```

```python
from sklearn import tree
model2 = tree.DecisionTreeClassifier()
model2.fit(x_train,y_train)
```

```python
model2.score(x_test,y_test)*85
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_pred)
cm
```

```python
import seaborn as sn
plt.figure(figsize = (10,7))
sn.heatmap(cm,annot=True)
plt.xlabel('predicted')
plt.ylabel('truth')
```

```python
df = pd.DataFrame(iris.data)
```

```python
df.head()
```

```python
df['target'] = iris.target
```

```python
df.head()
```

```python
iris.target_names
```

```python
x = df.drop(['target'],axis = 'columns')
y = df.target
```

```python
x_train2,x_test2,y_train2,y_test2 = train_test_split(x,y,test_size= 0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier
model3 = RandomForestClassifier(n_estimators=10)
model3.fit(x_train2,y_train2)
```

```python
model3.score(x_test2,y_test2)
```

```python
pred2 = model3.predict(x_test2)
```

```python
from sklearn.metrics import confusion_matrix
cm  =  confusion_matrix(y_test2,pred2)
cm
```

```python
plt.figure(figsize = (10,7))
sn.heatmap(cm, annot=True)
plt.title("confusion matrix")
plt.xlabel('predicted')
plt.ylabel('truth')
```
